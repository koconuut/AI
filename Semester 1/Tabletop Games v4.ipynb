# Question:
In Uno, what's the probability that the first card dealt is either a wild card or a blue reverse card? Assume no other cards have been dealt.
# Study Description:
I will be conducting an observational study because I will be observing data that is produced from multiple trials of an Uno simulation without changing any variables in the simulation. The sample space will be all the cards in an Uno deck, and there are 108 cards in an Uno deck. One trial of this simulation will be drawing a single random card from the deck. The outcome I am measuring is how often the first card dealt is either a wild card or a blue reverse card.

```python
import random
import seaborn as sns
import matplotlib.pyplot as plt

# --- Build Uno deck (108 cards: 8 wilds, 2 blue reverses, 98 others) ---
deck = ["Wild"] * 8 + ["Blue Reverse"] * 2 + ["Other"] * (108 - 10)

# --- Run 100 trials ---
results = []
for trial in range(1, 101):
    drawn = random.choice(deck)  # draw a single card
    outcome = "Wild or Blue Reverse" if drawn in ["Wild", "Blue Reverse"] else "Other"
    results.append(outcome)
    print(f"Trial {trial}: {drawn} -> {outcome}")

# --- Count frequencies ---
counts = {res: results.count(res) for res in set(results)}

# --- Print summary ---
print("\nSummary:")
print(f"Trials with Wild or Blue Reverse: {counts.get('Wild or Blue Reverse', 0)}")
print(f"Trials with Other: {counts.get('Other', 0)}")

# --- Visualization ---
sns.barplot(x=list(counts.keys()), y=list(counts.values()))
plt.title("Results of 100 Uno Trials (First Card Wild or Blue Reverse)")
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.show()
```

The theoretical probability of drawing either a wild card or a blue reverse card as the first card is calculated by adding the probabilities of each individual event. There are 4 wild cards and 2 blue reverse cards in an Uno deck, for a total of 6 favorable cards out of 108 cards. So, the theoretical probability is 6/108, which simplifies to 1/18 or approximately 5.56%.

To calculate the empirical probability, I divide the number of trials where a wild card or a blue reverse card was drawn by the total number of trials. In this simulation, 8 out of 100 trials resulted in either a wild card or a blue reverse card. Therefore, the empirical probability is 8/100, which is 8%.

Although the empirical probability is slightly higher than the theoretical probability, it's pretty close, showing that the simulation roughly aligns with the expected theoretical outcome for an Uno deck.
