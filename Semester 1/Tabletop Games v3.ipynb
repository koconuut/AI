# Question: 
In Uno, what's the probability of drawing 2 wild cards in a row from the deck? Assume no other cards have been dealt.
# Study Description: 
I will be conducting an observational study because I will be observing data that is produced from multiple trials of an Uno simulation without changing any variables in the simulation. The sample space will be all the cards in an Uno deck, and there are 108 cards in an Uno deck. One trial of this simulation will be drawing a random card. The outcome I am measuring is how often 2 wild cards in a row is drawn.

```python
import random
import seaborn as sns
import matplotlib.pyplot as plt

# --- Build Uno deck (108 cards: 8 wilds, 100 others) ---
deck = ["Wild"] * 8 + ["Other"] * (108 - 8)

# --- Run 100 trials ---
results = []
for trial in range(1, 101):
    drawn = random.sample(deck, 2)  # draw 2 cards without replacement
    outcome = "2 Wilds" if drawn[0] == "Wild" and drawn[1] == "Wild" else "Not 2 Wilds"
    results.append(outcome)
    print(f"Trial {trial}: {drawn} -> {outcome}")

# --- Count frequencies ---
counts = {res: results.count(res) for res in set(results)}

print("\nSummary:")
print(f"Trials with 2 Wilds: {counts.get('2 Wilds', 0)}")
print(f"Trials without 2 Wilds: {counts.get('Not 2 Wilds', 0)}")

# --- Visualization ---
sns.barplot(x=list(counts.keys()), y=list(counts.values()))
plt.title("Results of 100 Uno Trials (Drawing 2 Cards)")
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.show()
```

The theoretical probability of drawing 2 wild cards in a row from a full Uno deck is 8/108 times 7/107 which is around 0.00485, or around 0.5%. We got these numbers because there are 8 wild cards in a full Uno deck of 108 cards, and then after you draw a wild card the first time there's 7 wild cards remaining in the deck and 107 total cards because we already drew one card.

To calculate my empirical probability, I will divide the number of times I drew 2 wild cards in a row by the total number of trials I conducted. In this scenario, I drew 2 wild cards in a row 3 different times in the 100 trials. My empirical probability would be 3/100, or 3%.

This shows that my empirical probability is higher than the theoretical probability. We got very different answers because of complete randomness that happens in our simulation, and we got really lucky during our 100 trials and that led to a very different empirical probability compared to the theoretical probability.
