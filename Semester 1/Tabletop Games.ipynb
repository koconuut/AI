# Question: 
Given that the first card dealt in a game of Uno is blue, what is the probability that it's a reverse card?

# Study Description: 
I will be conducting an observational study because I will be observing data that is produced from multiple trials of an Uno simulation without changing any variables in the simulation. The sample space will be all the blue cards in an Uno deck, and there are 25 total blue cards in 1 Uno deck. One trial of this simulation will be drawing a random blue card. The outcome I am measuring is how often a blue reverse card is drawn.

```python
import random
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# --- Setup ---
blue_cards = ["reverse card"] * 2 + ["other"] * 23  # 25 total
trials = 100
results = []

# --- Run simulation ---
for trial in range(1, trials + 1):
    drawn_card = random.choice(blue_cards)
    results.append(drawn_card)
    print(f"Trial {trial}: {drawn_card}")  # print each trial result

# --- Convert to DataFrame ---
df = pd.DataFrame({"Trial": range(1, trials + 1), "Card": results})

# --- Count occurrences ---
reverse_count = results.count("reverse card")
other_count = results.count("other")

print("\nSummary:")
print(f"Reverse cards drawn: {reverse_count}")
print(f"Other blue cards drawn: {other_count}")

# --- Visualization ---
sns.countplot(x="Card", data=df, palette="Blues")
plt.title("Total Counts of Drawn Blue Cards (100 Trials)")
plt.xlabel("Card Type")
plt.ylabel("Count")
plt.show()
```

The theoretical probability of drawing a blue reverse card is 2/25 or 8%, because there are 2 blue reverse cards out of 25 total blue cards in an Uno deck. To calculate my empirical probability, I will divide the amount of reverse cards drawn by the amount of trials I conducted. In this scenario, I drew 8 reverse cards, and I drew cards 100 times in total. My empirical probability would be 8/100, which is also 8%, therefore my empirical probability and my theoretical probability are the same. This shows that what we assumed would happen with the theoretical probability is consistent with what ended up happening in our empirical probability, showing that they are both consistent with the Uno deck.
