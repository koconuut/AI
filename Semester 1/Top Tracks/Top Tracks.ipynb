```python
# Question: Which Pop artist was the most frequently observed on Spotify's U.S. Top Tracks of 2024 playlist?
# The study I will be conducting is an observational study, because I'm looking at existing data from the previous year to figure out the answer to my question.
# I will be examining a sample of the population that listens to Pop music.
# The data I collected originated from Mrs. Spindt and AI, she was able to hand-collect the data and figure out the different genres listed with the help of AI. Some biases that come with using AI include the fact that AI's information may not always be accurate when comparing the genres it considers to genres we consider.
```

```python
# imports + variables
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load CSV into dataframe
df = pd.read_csv("Top Tracks Playlists - USA.csv")

# Filter only pop artists
pop_artists = df[df["AI Genre"] == "Pop"]


# Create a frequency table of how many times each pop artist appears
frequency_table = pop_artists["Artist"].value_counts()

print(frequency_table)

# Count frequencies
frequency_df = pop_artists["Artist"].value_counts().reset_index()
frequency_df.columns = ["Artist", "Count"]


# Plot with seaborn
plt.figure(figsize=(10,6))  # adjust size if needed
sns.barplot(data=frequency_df, x="Artist", y="Count", palette="viridis")

# Make chart readable
plt.xticks(rotation=45, ha="right")
plt.title("Frequency of Pop Artists")
plt.xlabel("Artist")
plt.ylabel("Number of Songs")

plt.show()
```

```python
# relative frequency

pop_df = df[df["AI Genre"] == "Pop"]

pop_artist_counts = pop_df["Artist"].value_counts()

pop_rel_freq = pop_artist_counts / pop_artist_counts.sum()
print(pop_rel_freq)

sns.barplot(x=pop_rel_freq.index, y=pop_rel_freq.values, palette="pastel")
plt.title("Relative Frequency of Pop Artists")
plt.xticks(rotation=45, ha="right")
plt.xlabel("Artist")
plt.ylabel("Relative Frequency")
plt.show()
```

```python
# answer

print("Sabrina Carpenter was the most frequently observed pop artist on Spotify’s U.S. Top Tracks of 2024 playlist, with 4 of the 50 songs belonging to her.")
```
