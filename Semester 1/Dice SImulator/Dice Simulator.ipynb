# Dice Simulator

```python
import random
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# --- Step 1: Ask user for N-sided die ---
sides = int(input("Enter the number of sides on the die (N): "))

# --- Step 2: Run simulation ---
n_trials = 100
rolls = [random.randint(1, sides) for _ in range(n_trials)]

# --- Step 3: Convert results to a DataFrame ---
df = pd.DataFrame(rolls, columns=["Die Roll"])

# --- Step 4a: Bar chart of results ---
plt.figure(figsize=(8,6))
sns.countplot(x="Die Roll", data=df, palette="viridis", order=range(1, sides+1))
plt.title(f"Distribution of {n_trials} Rolls of a {sides}-Sided Die", fontsize=14)
plt.xlabel("Die Face")
plt.ylabel("Frequency")
plt.show()

# --- Step 5: Print rolls if you want to see them ---
print("All 100 rolls:")
print(rolls)
```
